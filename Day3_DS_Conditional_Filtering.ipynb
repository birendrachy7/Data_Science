# Conditional Filtering

Conditional Filtering means selecting only those rows from a DataFrame that satisfy a specific condition.

Think of it like applying a filter in Excel:

"Show me only students whose marks are greater than 80."

Instead of viewing all data, you view only the records that match your condition.

# Data Types in a DataFrame

A data type defines the kind of values that can be stored in a column.

### Integer (int)

Stores whole numbers without decimal points.

Examples:

1. Age
2. Number of students
3. Quantity sold

Characteristics:

1. Used for counting.
2. Takes less memory than floating-point numbers.
### Float (float)

Stores numerical values with decimal points.

Examples:

Salary
Height
Weight
Temperature

Characteristics:

Used when precision is required.
Represents continuous measurements.
### String (str or object)

Stores textual data.

Examples:

1. Name
2. Address
3. Country
4. Department

Characteristics:

1. Used for descriptive information.
2. Cannot be used directly for mathematical calculations.
### Boolean (bool)

Stores logical values.

Possible values:

1. True
2. False

Examples:

* Is Student?
* Is Active?
* Is Employed?

Characteristics:

* Frequently used in filtering and conditional operations.
### Datetime

Stores date and time information.

Examples:

* Birth Date
* Joining Date
* Transaction Date

Characteristics:

* Enables time-based analysis.
* Useful for trend and forecasting analysis.
### Complex Numbers

Stores numbers having real and imaginary parts.

Example:

* 3 + 4j

Characteristics:

* Mostly used in scientific and engineering computations.
* Rarely used in business datasets.
### Missing Values (NaN/None)

Represents unavailable or unknown information.

Examples:

* Name      Salary
* Ram       50000
* Hari      NaN

Characteristics:

* Indicates incomplete data.
* Requires special handling during data cleaning.

# Types of Data in a Dataset

Data can be broadly divided into two major categories.

# A. Numerical Data (Quantitative Data)

Numerical data represents measurable quantities and can be used for mathematical calculations.

Examples:

* Age
* Salary
* Marks
* Height

Numerical data is further divided into:

### Discrete Data

Discrete data consists of countable values, whole number, unique number, must be in range

Examples:

* Number of students in a class
* Number of cars owned
* Number of employees

Characteristics:

* Usually whole numbers.
* Cannot have intermediate values.

Example:

    Students = 25

You cannot have:

    Students = 25.5
### Continuous Data

Continuous data can take any value in range or not in range,can be whole number or decimal valuese.

Examples:

* Height
* Weight
* Temperature
* Time
* Price
* Sales
* Marks
* Salary

Characteristics:

* Can contain decimal values.
* Obtained through measurement.

Example:

    Height = 170.45 cm
# B. Categorical Data (Qualitative Data)

Categorical data represents categories or labels rather than numerical measurements.

Examples:

* Gender
* Department
* Country
* Product Type

Categorical data is divided into:

### Nominal Data

Nominal data consists of categories without any natural order.

Examples:
* Blood Group
* Color
* Gender
* Caste
* Race 

Characteristics:

* Categories are different but not ranked.
* Mathematical operations are not meaningful.
### Ordinal Data

Ordinal data consists of categories that have a meaningful order.

Examples:

* Grade
* Food

Characteristics:

* Ranking exists.
* The exact difference between categories is unknown.

Numeric case:
------------------
mean -> if our data is in range -> Best when no significant outliers exist.

median -> if our data is not in range -> Best when data contains outliers or is skewed.


Numeric and Categorical both case:

# Outliers

An outlier is an observation that differs significantly from the rest of the data.

### Definition:

A value that lies unusually far away from other observations in a dataset.

### Effects of Outliers

* Distort the mean.
* Increase variance.
* Affect statistical models.
* Lead to incorrect conclusions.

### Why Are Outliers Important?

Before performing analysis or building machine learning models, outliers must be identified because they can:

* Reduce model accuracy.
* Misrepresent trends.
* Influence predictions.

Common detection methods include:
* Box Plots
* Interquartile Range (IQR)
* Z-Score Method
