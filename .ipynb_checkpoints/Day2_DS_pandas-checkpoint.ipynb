# Pandas
Pandas is a Python library used for working with data sets.  
It has functions for analyzing, cleaning, exploring, and manipulating data.  
The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

### Why Use Pandas?
1. Pandas allows us to analyze big data and make conclusions based on statistical theories.

2. Pandas can clean messy data sets, and make them readable and relevant.

3. Relevant data is very important in data science.

# .loc[]
.loc[] is label-based indexing.  
It selects data using the row labels (index names) and column labels (column names).   
It is used to slice the data using alphabet index  
### Syntax
    df.loc[row_label, column_label]

```python
personal_details
```

```python
#Selecting the single row
personal_details.loc['A']
```

```python
# Selecting multiple rows
personal_details.loc['A':'D']
```

```python

# Selecting a Single Value
personal_details.loc['B':'D','Name']
```

```python
# Selecting multiple cloumns
personal_details.loc[:,['Name','Address']]  # : means all rows.
                                            # select only Name and Address
```

```python
# With .loc, the ending index is INCLUDED.
personal_details.loc['B':'D']
```

```python
# Boolean Filtering with .loc
personal_details.loc[personal_details['Age']>25]
```

```python
personal_details[['Name','Address']]  # for pandas
#select name, address form personal_details);   # for SQL
```

```python
personal_details.loc[
    ['B','D'],
    ["Name","Address"]
]
```

# .iloc[]
.iloc[] is integer position-based indexing.   
It selects data using the numerical position of rows and columns.   
ILoc -> Integer Location -> used to slice data using numeric index

### Syntax
    df.iloc[row_position, column_position]

```python
personal_details
```

```python
#Selecting first row
personal_details.iloc[0]
```

```python
personal_details.iloc[1]
```

```python
# selecting single value
personal_details.iloc[1,0]
```

```python
personal_details.iloc[1:3,0:4]
```

```python
personal_details.iloc[
        [1,3],
        [1,2]
    ]
```
