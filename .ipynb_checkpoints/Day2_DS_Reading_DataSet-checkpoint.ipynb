# What Does "Reading a Dataset" Mean?

A dataset is a collection of data stored in a file, database, or other source.

Before we can analyze data using Pandas, we must first load (read) the dataset into memory.

Pandas stores data mainly in a structure called a DataFrame.
### Data Flow
                Dataset File
                      ↓
                Pandas reads file
                      ↓
                DataFrame created
                      ↓
                Data Analysis

# Why Do We Read Datasets?

Data usually comes from:
1. CSV files
2. Excel files
3. JSON files
4. SQL databases
5. APIs
6. Text files
       

Pandas provides functions to read data from all these sources into a DataFrame.

Once loaded, you can:
1. Clean data
2. Filter records
3. Calculate statistics
4. Create visualizations
5. Build machine learning models

##### Function to connect to the different data files using pandas
1. read_csv()   
2. read_excel()
3. read_json() 
4. read_sql()

##### Syntax
    pd.read_csv("filename.csv")

```python
# Step 1: importing pandas
import pandas as pd
```

```python
#reading for csv File
emp=pd.read_csv('Dataset/Employees.csv')
emp
```

# .head()

The .head() method returns the first few rows of a DataFrame or Series.

It is mainly used to:

1. Preview the dataset after loading it.
2. Verify that data was read correctly.
3. Check column names and data values.
4. Understand the structure of the dataset.
#### Syntax
    data_file_name.head()

By default, it returns the first 5 rows.

```python
# to display top 5 record
emp.head()  # default 5 
```

```python
emp.head(15)
```

# .tail()
The .tail() method returns the last few rows of a DataFrame or Series.

It is useful for:

1. Checking the end of a dataset.
2. Verifying that all records were loaded.
3. Inspecting recent entries in time-series data.
#### Syntax
    df.tail()

By default, it returns the last 5 rows.

```python
emp[:5]  # slicing 
```

```python
# to display bottom 5
emp.tail()
```

```python
emp.tail(10)
```

```python
emp.info() # to see the information about the file
```
